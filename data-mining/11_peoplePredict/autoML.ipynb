# autoML-没有做特征工程

```python
from autogluon.tabular import TabularDataset, TabularPredictor
import pandas as pd
```

```python
from operator import index

df_train = TabularDataset('assets/dataTrain.csv')
df_test = TabularDataset('assets/dataA.csv')
```

```python
# df_train.set_index(["id"], inplace=True)
# df_test.set_index(["id"], inplace=True)
```

```python
df_train
```

```python
df_test
```

```python
# 对缺失值进行填充
df_train.fillna(-1, inplace=True)
df_test.fillna(-1, inplace=True)

df_train.isnull().sum()
```

```python
df_test
```

```python
predictor = TabularPredictor(label='label').fit(train_data=df_train,presets='best_quality')
```

```python
predictions = predictor.predict(df_test)
```

```python
df_test['label'] = predictions
```

```python
df_test[['id', 'label']].to_csv('submit/autoML-无特征工程.csv', index=False, encoding='utf-8')
```
